```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
len(df)
```

```python
df.dtypes
```

```python
df.rename(columns={'YearsExperience': 'Experience'}, inplace=True)
```

```python
df.head()
```

```python
x_axes = df.Experience
y_axes = df.Salary
```

```python
# plotting first scatterplot
plt.scatter(x_axes, y_axes, color='blue')
plt.xlabel('Experience (years)')
plt.ylabel('Salary (usd)')
plt.show()
plt.clf()
```

```python
# create our model
model = sm.OLS.from_formula('Salary ~ Experience', data=df)
results = model.fit()
y_intercept, slope = results.params
```

```python
# y_prediction
y_prediction = (slope * x_axes) + y_intercept
```

```python
# create a scatterplot but with line
plt.scatter(x_axes, y_axes, color='blue')
plt.plot(x_axes, y_prediction, color='red')
plt.xlabel('Experience (years)')
plt.ylabel('Salary (usd)')
plt.show()
plt.clf()
```

```python
# fitted_values
fitted_values = results.predict(df)

# residuals
residuals = y_axes - fitted_values
```

```python
# running tests
# normal assumption - histogram of the residuals
plt.hist(residuals)
plt.show()
plt.clf()
```

```python
# Homoscedasticity assumption
# scatterplot of fitted values and residuals
plt.scatter(fitted_values, residuals)
plt.show()
plt.clf()
```

```python
# final function
predict_salary = lambda x: results.predict({'Experience': x})[0].round(2)
```

```python
# examples
predict_salary(2)
```

```python
predict_salary(1.2)
```

```python
predict_salary(22)
```
